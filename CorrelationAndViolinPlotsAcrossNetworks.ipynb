```python
dfs = [
    df_bio_yeastcollins,
    df_bio_yeastprotein,
    df_infra_euairlines,
    df_infra_usairports,
    df_infra_uspowergrid,
    df_humansocial_adolescenthealth,
    df_osn_caltech,
    df_osn_deezereu,
    df_osn_egofacebook,
    df_osn_facebookfriends,
    df_osn_facebookorganizations,
    df_osn_facebookpoliticianpages,
    df_osn_hamsterster,
    df_osn_princeton,
    df_misc_dncemails
]


len(dfs)
```

## Mean 

```python
df_mean_across_networks = pd.concat([each.stack() for each in dfs],axis=1).apply(lambda x:x.mean(),axis=1).unstack()
df_mean_across_networks
```

```python
fig, ax = plt.subplots(figsize=(6, 3.5)) 
mask = np.zeros_like(df_mean_across_networks)
mask[np.triu_indices_from(mask)] = 1
sns.heatmap(df_mean_across_networks, mask= mask, ax= ax, annot=True, cmap="seismic_r")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 3.5)) 
mask = np.zeros_like(df_mean_across_networks)
mask[np.triu_indices_from(mask)] = 1
sns.heatmap(df_mean_across_networks, mask= mask, ax= ax, vmin=-1, vmax=1, annot=True, cmap="seismic_r")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
ax = sns.violinplot(data=df_mean_across_networks)
plt.xticks(rotation='vertical')
plt.tick_params(labelsize=13)
```
